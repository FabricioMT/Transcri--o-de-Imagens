```python
import cv2
import pytesseract
from os import listdir
pytesseract.pytesseract.tesseract_cmd = r'C:\\Users\\fabri\\AppData\\Local\\Programs\\Tesseract-OCR\\tesseract.exe'


def load_images():
    file_names = listdir('./targets/')
    targets = []
    count = 0

    for file in file_names:
        count += 1
        path = 'targets/' + file
        targets = cv2.imread(path)
        text = pytesseract.image_to_string(targets)
        print(text)
        print("----------------------------------------")
        print(f" End Page {count} ")
        print("----------------------------------------")

load_images()
```
